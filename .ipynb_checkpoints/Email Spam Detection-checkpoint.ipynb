# Email Spam Detection using Ensemble Methods

#### 1. Importing required Libraries

```python
# Essential libraries for data manipulation and machine learning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
import nltk
from nltk.corpus import stopwords
import re
```

#### 2. Read csv file

```python
df=pd.read_csv("spam.csv",encoding='latin-1')
```

#### 3. Text Preprocessing & Exploratory Data Analysis

```python
df.head(5)
```

```python
# Keep only relevant columns and rename them
df = df.rename(columns={'v1': 'label', 'v2': 'text'})
df = df[['label', 'text']]
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Analyze the distribution of spam vs ham messages
class_distribution = df['label'].value_counts()
print("Class Distribution:")
print(class_distribution)
print("\nPercentage Distribution:")
print(df['label'].value_counts(normalize=True) * 100)

# Visualize the distribution

plt.figure(figsize=(8, 6))
class_distribution.plot(kind='bar')
plt.title('Distribution of Spam vs Ham Messages')
plt.xlabel('Message Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
def preprocess_text(text):
    """
    Clean and preprocess text data:
    1. Convert to lowercase
    2. Remove special characters and digits
    3. Remove extra whitespace
    4. Remove stopwords
    """
    # Convert to lowercase
    text = text.lower()
    # Remove special characters and digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text


df['cleaned_text'] = df['text'].apply(preprocess_text)
```

#### Message Length Analysis

```python
# Add message length as a feature
df['message_length'] = df['cleaned_text'].str.len()

# Analyze message lengths
print("Message Length Statistics:")
print(df.groupby('label')['message_length'].describe())

# Visualize message lengths by class
plt.figure(figsize=(10, 6))
plt.hist(df[df['label']=='ham']['message_length'], 
         bins=50, alpha=0.5, label='Ham', color='blue')
plt.hist(df[df['label']=='spam']['message_length'], 
         bins=50, alpha=0.5, label='Spam', color='red')
plt.title('Distribution of Message Lengths by Class')
plt.xlabel('Message Length')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#### Common words Analysis

```python
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize

# Function to get most common words
def get_common_words(text_series, n=10):
    all_words = ' '.join(text_series).lower()
    word_tokens = word_tokenize(all_words)
    # Remove single-character words and stopwords
    words = [word for word in word_tokens if len(word) > 1]
    return Counter(words).most_common(n)

spam_common = get_common_words(df[df['label'] == 'spam']['cleaned_text'])
ham_common = get_common_words(df[df['label'] == 'ham']['cleaned_text'])
# Convert to DataFrame for easier plotting
spam_df = pd.DataFrame(spam_common, columns=['Word1', 'Count'])
ham_df = pd.DataFrame(ham_common, columns=['Word2', 'Count'])
# Create bar plots
fig, axes = plt.subplots(2, 1, figsize=(14, 6))

# Spam message common words
axes[0].barh(spam_df['Word1'], spam_df['Count'], color='red')
axes[0].set_title('Most Common Words in SPAM Messages')
axes[0].invert_yaxis()  # Invert y-axis for better readability
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Word')

# Ham message common words
axes[1].barh(ham_df['Word2'], ham_df['Count'], color='blue')
axes[1].set_title('Most Common Words in HAM Messages')
axes[1].invert_yaxis()  # Invert y-axis for better readability
axes[1].set_xlabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```

#### 5. TF-IDF Feature Extraction

TF-IDF (Term Frequency-Inverse Document Frequency) is a way to convert text into numbers that machine learning models can understand:

- TF (Term Frequency): How often a word appears in a message
- IDF (Inverse Document Frequency): How unique or rare that word is across all messages
- If a word appears a lot in one message (high TF) but is rare across all messages (high IDF), it's probably important!

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF vectorizer
# max_features=3000 means we'll keep only the 3000 most important words
tfidf = TfidfVectorizer(max_features=3000)

# Convert text data to TF-IDF features
X = tfidf.fit_transform(df['cleaned_text'])
y = df['label'].map({'spam': 1, 'ham': 0})  # Convert labels to 0 and 1

# Show the shape of our feature matrix
print("Feature matrix shape:", X.shape)

# Look at some feature names (words)
print("\nSample features (words):", tfidf.get_feature_names_out()[:10])
```

#### Splitting data into train and test

```python
from sklearn.model_selection import train_test_split

# Split the data: 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,    # 20% for testing
    random_state=42,  # For reproducibility
    stratify=y        # Maintain class distribution in both sets
)

# Print the shapes of our training and testing sets
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
print("\nClass distribution in training set:")
print(pd.Series(y_train).value_counts(normalize=True))
print("\nClass distribution in test set:")
print(pd.Series(y_test).value_counts(normalize=True))
```

#### Training and Evaluating models

##### 1. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
# Train the model
rf_model.fit(X_train, y_train)
# Make predictions
rf_y_pred = rf_model.predict(X_test)

# Print classification report
print(f"\nRandom Forest Classification Report:")
print(classification_report(y_test, rf_y_pred))
```

##### 2. Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
# Train the model
gb_model.fit(X_train, y_train)
# Make predictions
gb_y_pred = gb_model.predict(X_test)
# Print classification report
print(f"\nGradient Boosting Classification Report:")
print(classification_report(y_test, gb_y_pred))
```

#### Random Forest VS Gradient Boosting Confusion matrix 

```python
fig, axes = plt.subplots(1,2, figsize=(14,6))

labels=["Ham","Spam"]
# Random forest Confusion matrix
rf_cm = confusion_matrix(y_test, rf_y_pred)
sns.heatmap(rf_cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],xticklabels=labels, yticklabels=labels)
axes[0].set_title('Random Forest Confusion Matrix')
axes[0].set_ylabel('True Label')
axes[0].set_xlabel('Predicted Label')

# Plot Gradient Boosting Confusion Matrix
gb_cm = confusion_matrix(y_test, gb_y_pred)
sns.heatmap(gb_cm, annot=True, fmt='d', cmap='Blues', ax=axes[1],xticklabels=labels, yticklabels=labels)
axes[1].set_title('Gradient Boosting Confusion Matrix')
axes[1].set_ylabel('True Label')
axes[1].set_xlabel('Predicted Label')

plt.tight_layout()
plt.show()
```

#### Implementing Voting Classifier

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# Initialize base models
log_reg = LogisticRegression(random_state=42)
dt_clf = DecisionTreeClassifier(random_state=42)
svm_clf = SVC(probability=True, random_state=42)

# Create voting classifier
# Can use either 'hard' voting (majority vote) or 'soft' voting (probability-based)
voting_clf = VotingClassifier(
    estimators=[
        ('lr', log_reg),
        ('dt', dt_clf),
        ('svc', svm_clf)
    ],
    voting='soft'
)

# Train the voting classifier
print("Training Voting Classifier...")
voting_clf.fit(X_train, y_train)

# Make predictions
voting_y_pred = voting_clf.predict(X_test)

# Print classification report
print("\nVoting Classifier Results:")
print(classification_report(y_test, voting_y_pred))

# Plot confusion matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, voting_y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Voting Classifier Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

models = {
    'Logistic Regression': log_reg,
    'Decision Tree': dt_clf,
    'SVM': svm_clf,
    'Voting Classifier': voting_clf
}

# Compare all models
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    # Get predictions
    each_y_pred = model.predict(X_test)
    
    # Store results
    results[name] = {
        'Accuracy': accuracy_score(y_test, each_y_pred),
        'Precision': precision_score(y_test, each_y_pred),
        'Recall': recall_score(y_test, each_y_pred),
        'F1': f1_score(y_test, each_y_pred)
    }

# Create comparison DataFrame
results_df = pd.DataFrame(results).T
print("\nModel Comparison:")
print(results_df)

# Plot comparison
plt.figure(figsize=(12, 6))
results_df.plot(kind='bar')
plt.title('Model Performance Comparison')
plt.xlabel('Model')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
